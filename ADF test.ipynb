```python
#importing libraries
import pandas as pd 
import matplotlib.pyplot as plt 
from statsmodels.tsa.stattools import adfuller

#reading price data
prices_df = pd.read_csv('AUDCAD.csv',index_col=0)

## Plotting
prices_df.prices.plot(figsize = (12,4))
plt.show()

## Perform ADF Test

#use help function to know more about adfuller object
help(adfuller)

adf = adfuller(prices_df.prices, maxlag=1)

##  ADF Test statistics

current_adf = adf[0]

## Critical values

adf_critearia = adf[4]

# -3.19 < -2.56 so we are having 95% confident of this price series being stationary
print(current_adf)
print(adf_critearia)
```
